```python
from sklearn.datasets import make_swiss_roll
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
from torchvision import datasets, transforms

train_dataset = datasets.MNIST(root='./mnist_data/', train=True, transform=transforms.ToTensor(), download=True)

mnist_train = train_dataset.data.numpy().copy()
mnist_train = mnist_train.reshape(60000,-1)
```

```python
training_size = 1000

mnist_train_targets = train_dataset.targets.numpy().copy()

Y = mnist_train_targets[0:training_size]

color_list = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']

custom_color_map = [color_list[y_val] for y_val in Y]
```

```python
import numpy as np
from sklearn.manifold import TSNE

X_embedded = TSNE(n_components=3,perplexity=10,n_iter=1000,learning_rate=1000).fit_transform(mnist_train[0:training_size,:])

X_embedded.shape

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

xs = X_embedded[:,0];
ys = X_embedded[:,1];
zs = X_embedded[:,2];
ax.scatter(xs, ys,zs, alpha=0.6,c=custom_color_map, edgecolors='w')

ax.set_xlabel('x_axis')
ax.set_ylabel('y_axis')

plt.show()
```
