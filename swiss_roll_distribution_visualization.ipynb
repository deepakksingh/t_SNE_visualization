```python
from sklearn.datasets import make_swiss_roll
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
sr_data_points,t = make_swiss_roll(n_samples=1000, random_state=123)
```

```python
# sr_data_points
```

```python
x_val = sr_data_points[:,0]
# x_val
```

```python
y_val = sr_data_points[:,1]
# y_val
```

```python
z_val = sr_data_points[:,2]
# z_val
```

```python
# from matplotlib import interactive
# interactive(True)

%matplotlib inline
# %matplotlib notebook
# %matplotlib ipympl
# %matplotlib nbAgg
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

xs = x_val;
ys = y_val
zs = z_val
ax.scatter(xs, ys, zs, s=50, alpha=0.6, edgecolors='w')

ax.set_xlabel('x_axis')
ax.set_ylabel('y_axis')
ax.set_zlabel('z_axis')

plt.show()
```

```python
import numpy as np
from sklearn.manifold import TSNE

X_embedded = TSNE(n_components=2).fit_transform(sr_data_points)

X_embedded.shape

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111)

xs = X_embedded[:,0];
ys = X_embedded[:,1];
ax.scatter(xs, ys, alpha=0.6, edgecolors='w')

ax.set_xlabel('x_axis')
ax.set_ylabel('y_axis')

plt.show()
```
